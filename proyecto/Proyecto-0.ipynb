# 1. Introducción

El problema de N-Cuerpos fue planteado por Isaac Newton (1687) en su libro *Principios matemáticos de la filosofía natural*. La finalidad de este problema es modelar el movimiento de N partículas, movidas por fuerzas gravitacionales mediante ecuaciones diferenciales ordinarias no lineales. 

Para dos cuerpos es fácil predecir su posición y velocidad en cualquier instante, mientras que para 3 o más cuerpos el sistema no es soluble analíticamente. Newton descubrió que la trayectoria de un cuerpo, sujeto a una fuerza central varía como el inverso cuadrado de la distancia: *la ley de gravitación universal*.

Este es un problema clásico de la mecánica celeste y la astronomía. El Sistema Solar es un ejemplo clásico donde el cambio de posición de los planetas al moverse alrededor del Sol resulta en una variación de la fuerza ejercida sobre alguno de sus miembros.

El desarrollo de las computadoras en la segunda mitad de este siglo ha tenido un impacto profundo en el problema de N-cuerpos tanto en el uso de programas con álgebra simbólica, como en la solución a través de simulaciones numéricas.

# 2. Solución propuesta

La solución sigue la siguiente sucesión:
* Se parte de una sola partícula. Se considera primeramente un universo euclídeo por lo que la partícula queda perfectamente definida por una posición en $x$, una posición en $y$, una velocidad en $x$, una velocidad en $y$ y su masa. A partir de las ecuaciones newtonianas de movimiento se calcula la posición de la partícula.
* Se emplea la *Segunda ley de Newton* para acelerar la particula a partir de su masa y *una* fuerza externa, definida por una fuerza en $x$ y una fuerza en $y$.
* Se considera el caso cuando se aplican a la partícula diferentes fuerzas en distintos tiempos.
* Se resuelve el problema de calcular la distancia entre dos partículas.
* A partir del punto anterior, se codifica la *Ley de gravitación universal* también de Newton.
* Estamos preparados para mover $N$ partículas...
* Utilizando el *principio de superposición* se calcula la fuerza que ejerce el sistema sobre cada partícula en el tiempo $t=0$ (la posición inicial).
* Estas fuerzas se aplican al sistema en el tiempo $t=1$.
* Se calcula la fuerza que ejerce el sistema sobre cada partícula en el tiempo $t=i$, $i>0$.
* Se aplican las fuerzas calculadas en el tiempo $t=i$, en el tiempo $t=i+1$, $i>0$.
* Se consideran posibles errores.
* Se considera qué hacen las partículas cuando llegan a los límites del universo.

Comenzamos simulando el movimiento de una partícula al aplicarle una fuerza dada durante un tiempo específico. A partir de la fuerza $F$ y de la masa del objeto $m$ , podemos calcular la magnitud $a$ con la que acelera a través de la **Segunda ley de Newton**:

$$
F = m a
$$

Una vez determinada la aceleración podemos aplicar las fórmulas vistas en secundaria:

$$
v_f = v_i t  + a t
$$

$$
d_f = v_i t + \frac{1}{2} a t^2 + d_i
$$

donde $v_f$ es la velocidad final del objeto, $v_i$ es la velocidad inicial del objeto, $t$ el tiempo, $a$ la aceleración, $d_f$ la posición final y  $d_i$ la posición inicial.

### 2.1.1 Primera ley de Newton

Según wikipedia :

*"Todo cuerpo persevera en su estado de reposo o movimiento uniforme y rectilíneo a no ser que sea obligado a cambiar su estado por fuerzas impresas sobre él."*

Observemos que nuestro código cumple con esta ley.

#### 2.1.1.1 Velocidad nula

De acuerdo a la primera ley de Newton, si nuestra `particula` está en reposo, deberá permanecer así si no se le aplica alguna fuerza.

¿Nuestro código cumple tal afirmación? Veamos que sí.

```python
posicion_reposo = np.array([0., 0.]) #Iniciamos nuestra partícula en el origen
velocidad_reposo = np.array([0., 0.]) #La partícula está en reposo
masa_reposo = 10.
fuerza_reposo = np.array([0., 0.]) #No aplicaremos fuerza
time = 83. #Simularemos durante 83 segundos
steps = 1000 
```

```python
particula_reposo = Particula(posicion = posicion_reposo, velocidad = velocidad_reposo, masa = masa_reposo)
```

```python
#Movemos la partícula
particula_reposo.mover(fuerza = fuerza_reposo, tiempo = time, pasos = steps)
```

```python
#Dibujamos la partícula
particula_reposo.dibujar()
```

Y nuestra partícula permaneció en reposo.

#### 2.1.2.2 ¿En qué nos ayuda `steps`?

`Steps` define mejor la trayectoria gráfica de la partícula. Veámoslo con el ejemplo anterior:

```python
posicion_0 = np.array([0.0, 0.0]) #NUestra partícula partirá del origen
velocidad_0 = np.array([-780.0, 33.3])
masa_0 = 10.
fuerza_0 = np.array([200.0, -50.0])
time = 83.0
#Función para cambiar el número de pasos
def moviendo_pasos(f, t, p):
    """
    f es la fuerza, t el tiempo y p los pasos
    """
    posicion_0 = np.array([0.0, 0.0]) #NUestra partícula partirá del origen
    velocidad_0 = np.array([-780.0, 33.3])
    masa_0 = 10.
    fuerza_0 = np.array([200.0, -50.0])
    time = 83.0
    particula_0 = Particula(posicion = posicion_0, velocidad = velocidad_0, masa = masa_0)
    particula_0.mover(fuerza = f, tiempo = t, pasos = p)
    particula_0.dibujar_contra_tiempo()
```

```python
interact(moviendo_pasos, f = fixed(fuerza_0), t = fixed(time), p = (0,100))
```

Observamos que con 0 pasos, la partícula parece no moverse; con 1 paso, sólo se dibuja la recta entre la posición final y la inicial. En cambio, con 100 pasos, se describe mejor la trayectoria de la partícula.

**Observación:** El mejor número de `pasos` depende del tiempo. Para 83 segundos son suficientes 100. Aún así *se recomiendan dos pasos por cada segundo*.

#### 2.1.2.3 Más fuerzas

Resuelto el problema de mover una partícula aplicándole una fuerza durante cierto tiempo, la pregunta natural es ¿cómo aplicar más de una fuerza en distintos tiempos?

##### 2.1.2.3.1 Primera solución

La primera solución que proponemos viene dada por el método `mover_muchas_fuerzas`:

```python
posicion_fuerzas = np.array([0., 0.]) #Nuestra partícula partirá del origen
velocidad_fuerzas = np.array([-20., 50.])
masa_fuerzas = 10.
```

```python
#Inicializamos nuestra partícula de prueba
particula_fuerzas = Particula(posicion = posicion_fuerzas, velocidad = velocidad_fuerzas, masa = masa_fuerzas)
```

```python
#El método recibe una lista con las fuerzas que se aplicarán
forces = [[-34.4,100.0], [23.65,-300.4], [2000.0,-10.0], [-30.0,80.0], [-2.90, 17.0], [200.0, -200.0]]
```

```python
#El método recibe una lista de tiempos cuya posición i determinará cuánto tiempo
#se aplica la fuerza i
#Donde acaba el tiempo anterior, sigue el otro
times = [[0.0, 5.5], [5.5, 8.9], [8.9, 30.0], [30.0, 45.4], [45.4, 45.6], [45.6, 74.3]]
```

```python
#EL método recibe una lista de steps cuya posición i determinará cuantos pasos
#se aplicarán a la fuerza i y tiempo i
#De acuerdo a lo analizado con anterioridad se requieren al menos dos pasos por segundo
steps = [12, 8, 44, 32, 2, 58]
```

```python
#Movemos la partícula
particula_fuerzas.mover_muchas_fuerzas(vector_tiempos = times, vector_fuerzas = forces, vector_pasos = steps)
```

```python
#Dibujamos el resultado
particula_fuerzas.dibujar_muchas_fuerzas()
```

##### 2.1.2.3.2 Segunda solución

La segunda solución aplica el método mover varias veces. Veamos el mismo ejemplo que antes:

```python
posicion_fuerzas_2 = np.array([0., 0.]) #Nuestra partícula partirá del origen
velocidad_fuerzas_2 = np.array([-20., 50.])
masa_fuerzas_2 = 10.
```

```python
#Inicializamos nuestra partícula de prueba
particula_fuerzas_2 = Particula(posicion = posicion_fuerzas_2, velocidad = velocidad_fuerzas_2, masa = masa_fuerzas_2)
```

```python
#Aplicamos 6 veces el método mover, 1 por cada fuerza 
particula_fuerzas_2.mover(fuerza = [-34.4, 100.0], tiempo = 5.5 - 0.0, pasos = 12)
particula_fuerzas_2.mover(fuerza = [23.65, -300.4], tiempo = 8.9 - 5.5, pasos = 8)
particula_fuerzas_2.mover(fuerza = [2000.0, -10.0], tiempo = 30.0 - 8.9, pasos = 44)
particula_fuerzas_2.mover(fuerza = [-30.0, 80.0], tiempo = 45.4 - 30.0, pasos = 32)
particula_fuerzas_2.mover(fuerza = [-2.90, 17.0], tiempo = 45.6 - 45.4, pasos = 2)
particula_fuerzas_2.mover(fuerza = [200.0, -200.0], tiempo = 74.3 - 45.6, pasos = 58)
```

```python
#Graficamos el resultado
particula_fuerzas_2.dibujar()
```

Observamos que se define exactamente la misma trayectoria.

##### 2.1.2.3.3 ¿Cuál es mejor?

Ambos métodos son en esencia el mismo. El método `mover_muchas_fuerzas` tiene la ventaja de que podemos ver las diferentes fuerzas gráficamente mientras que con el método `mover` únicamente la trayectoria continua. No obstante, en términos de practicidad es más eficiente `mover` pues para el otro método debemos conocer la fuerza externamente y aplicarla; se verá más adelante por qué esto es importante.
