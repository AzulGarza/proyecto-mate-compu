```python
import numpy as np
import matplotlib.pyplot as plt

class Particula:
    
    """
    Simula el movimiento de una partícula con atributos:
    -posicion: Es una lista o arreglo de posiciones iniciales [pos_x, pos_y] (números flotantes)
    -velocidad: Es una lista o arrgelo de velocidades iniciales [vel_x, vel_y] (números flotantes)
    -masa: Es una magintud (flotante)
    """
    
    def __init__(self, posicion, velocidad, masa):
        self.masa = masa
        
        self.posicion_0 = posicion
        
        self.velocidad_0 = velocidad
        
        self.posicion_actual = self.posicion_0

        self.G = 6.67384e-11
        
        self.velocidad_x = []
        self.velocidad_y = []
        
        self.velocidad_x.append(self.velocidad_0[0])
        self.velocidad_y.append(self.velocidad_0[1])
        
        self.posicion_x = []
        self.posicion_y = []
        
        self.posicion_x.append(self.posicion_0[0])
        self.posicion_y.append(self.posicion_0[1])
        
        self.tiempo = [0]
        self.movimientos = 0 #Lleva la cuenta del número de veces que se ha movido la partícula a través de pasos
        
        self.scatter_x = []
        self.scatter_y = []
        self.scatter_tiempo = []
        
 
    def mover(self, fuerza, tiempo, pasos):
        
        """
        La simulación per sé. Atributos:
        -fuerza: Una lista o arreglo [fuerza_x, fuerza_y]
        -tiempo: Una magnitud de tiempo
        -pasos: En cuántos intervalos se dividirá el tiempo (se recomiendan 100)
        
        Funcionamiento:
        Se divide el tiempo en pasos subintervalos. Para cada subintervalo se calcula la posición y la velocidad final conforme
        a la aceleración.
        Nota: los pasos definen mejor la trayectoria que dibuja la partícula.
        """
        
        self.pasos = pasos
        
        tiempo_aux = [0]
        
        for i in range(self.pasos):
            tiempo_aux.append((i+1)*(tiempo/self.pasos))
            self.tiempo.append((i+1)*(tiempo/self.pasos) + self.tiempo[self.movimientos])

        
        self.aceleracion_x = fuerza[0] / self.masa
        self.aceleracion_y = fuerza[1] / self.masa
        
        
        for i in range(self.pasos):
            self.velocidad_x.append(self.velocidad_x[self.movimientos] + tiempo_aux[i+1] * self.aceleracion_x)
            self.posicion_x.append(self.velocidad_x[self.movimientos] * tiempo_aux[i+1] + (0.5) * self.aceleracion_x * (tiempo_aux[i+1])**2
                                   + self.posicion_x[self.movimientos])
              
            self.velocidad_y.append(self.velocidad_y[self.movimientos] + tiempo_aux[i+1] * self.aceleracion_y)
            self.posicion_y.append(self.velocidad_y[self.movimientos] * tiempo_aux[i+1] + (0.5) * self.aceleracion_y * (tiempo_aux[i+1])**2
                                   + self.posicion_y[self.movimientos])
    
        self.movimientos += self.pasos
        
        self.posicion_actual[0] = self.posicion_x[self.movimientos]
        self.posicion_actual[1] = self.posicion_y[self.movimientos]
        
    
    def movimiento_puntual(self):
        
        """
        Divide el movimiento de la partícula para apreciar mejor cómo acelera.
        """
        
        if int(np.ceil(len(self.posicion_x)/10)) > 0:
            for i in range(0, len(self.posicion_x), int(np.ceil(len(self.posicion_x)/10))):
                self.scatter_x.append(self.posicion_x[i])
                self.scatter_y.append(self.posicion_y[i])
        else:
            for i in range(0, len(self.posicion_x)):
                self.scatter_x.append(self.posicion_x[i])
                self.scatter_y.append(self.posicion_y[i])  
        
        self.scatter_x.append(self.posicion_x[len(self.posicion_x)-1])
        self.scatter_y.append(self.posicion_y[len(self.posicion_y)-1])
        
    
    def dibujar(self):
        
        self.movimiento_puntual()
        
        plt.title(u"Posición Y vs Posición X", fontsize = 20)
        plt.plot(self.posicion_x, self.posicion_y , label="Movimiento", color="blue", linewidth = 1)
        plt.scatter(self.scatter_x, self.scatter_y, 30,  label = "Movimiento puntual", color = 'r')
        plt.plot(self.scatter_x[0], self.scatter_y[0], 'ko', label = "Pos inicial" )
        plt.plot(self.scatter_x[len(self.scatter_x)-1], self.scatter_y[len(self.scatter_y)-1], 'co', label = "Pos final" )
        plt.ylabel("Posicion Y (m)")
        plt.xlabel("Posicion X (m)")
        plt.legend(loc = 'best')

    def dibujar_xt_yt(self):
        
        fig, ax = plt.subplots(1,3, figsize = (19,5), dpi = 80)
        
        scatter_x = []
        scatter_y = []
        scatter_tiempo = []
        
        if int(np.ceil(len(self.posicion_x)/10)) > 0:
            for i in range(0, len(self.posicion_x), int(np.ceil(len(self.posicion_x)/10))):
                scatter_x.append(self.posicion_x[i])
                scatter_y.append(self.posicion_y[i])
                scatter_tiempo.append(self.tiempo[i])
        else:
            for i in range(0, len(self.posicion_x)):
                scatter_x.append(self.posicion_x[i])
                scatter_y.append(self.posicion_y[i])
                scatter_tiempo.append(self.tiempo[i])
        
        scatter_x.append(self.posicion_x[len(self.posicion_x)-1])
        scatter_y.append(self.posicion_y[len(self.posicion_y)-1])
        scatter_tiempo.append(self.tiempo[len(self.tiempo)-1])
        
        ax[0].set_title(u"Posición Y vs Posición X", fontsize = 20)
        ax[0].plot(self.posicion_x, self.posicion_y , label="Movimiento", color="blue", linewidth = 1)
        ax[0].scatter(scatter_x, scatter_y, 30,  label = "Movimiento puntual", color = 'r')
        ax[0].plot(scatter_x[0], scatter_y[0], 'ko', label = "Pos inicial" )
        ax[0].plot(scatter_x[len(scatter_x)-1], scatter_y[len(scatter_y)-1], 'co', label = "Pos final" )
        ax[0].set_ylabel("Posicion Y (m)")
        ax[0].set_xlabel("Posicion X (m)")
        ax[0].legend(loc = 'best')
        
        ax[1].set_title(u"Posición X vs Tiempo", fontsize = 20)
        ax[1].plot(self.tiempo, self.posicion_x , label=u"Movimiento uniforme", color='c')
        ax[1].scatter(scatter_tiempo, scatter_x, 30,  label = "Movimiento puntual", color = 'm')
        ax[1].plot(scatter_tiempo[0], scatter_x[0], 'ko', label = "Pos inicial" )
        ax[1].plot(scatter_tiempo[len(scatter_tiempo)-1], scatter_x[len(scatter_x)-1], 'yo', label = "Pos final" )
        ax[1].set_xlabel("Tiempo (s)")
        ax[1].set_ylabel("Posicion X (m)")
        ax[1].legend(loc = 'best')
        
        ax[2].set_title(u"Posición Y vs Tiempo", fontsize = 20)
        ax[2].plot(self.tiempo, self.posicion_y , label=u"Movimiento uniforme", color='g')
        ax[2].scatter(scatter_tiempo, scatter_y, 30,  label = "Movimiento puntual", color = 'k')
        ax[2].plot(scatter_tiempo[0], scatter_y[0], 'ro', label = "Pos inicial" )
        ax[2].plot(scatter_tiempo[len(scatter_tiempo)-1], scatter_y[len(scatter_y)-1], 'bo', label = "Pos final" )
        ax[2].set_xlabel("Tiempo (s)")
        ax[2].set_ylabel("Posicion Y (m)")
        ax[2].legend(loc = 'best')
        
    def distancia(self, otra_particula):
        
        self.norma = np.linalg.norm(np.array(self.posicion_actual) - np.array(otra_particula.posicion_actual)) # Interesa sólo la posición inicial
        self.vector_unitario = -1.0 * (np.array(self.posicion_actual) - np.array(otra_particula.posicion_actual))/self.norma
        
        return (self.norma, self.vector_unitario) # La distancia regresa la norma entre las dos posiciones y su vector unitario
    

    def fuerzaAplicada(self, otra_particula):
        
        """
        Calcula la fuerza que ejerce otra partícula sobre esta partícula.
        """
        
        d, vector_u = self.distancia(otra_particula)
        if d == 0:
            return np.array([0.0, 0.0])
        else:
            fuerza_gravitacional = (1/d) * vector_u * self.G * self.masa * otra_particula.masa
            return fuerza_gravitacional
        
    def mover_muchas_fuerzas(self, vector_tiempos, vector_fuerzas, vector_pasos):
        
        """
        -Tiempos es un vector de tiempos [[tiempo inicial, tiempo final], []]
        -Fuerzas es un vector de fuerzas
        -Paso es un vector de pasos
        En el lapso de tiempo i se aplicará la fuerza i con un número de p = paso
        con el método mover definido arriba.
        """
        
        self.vector_fuerzas = vector_fuerzas
        self.vector_pasos = vector_pasos
    
        
        for i in range(len(vector_fuerzas)):
            dt = vector_tiempos[i][1] - vector_tiempos[i][0]
            self.mover(vector_fuerzas[i], dt, vector_pasos[i])
    
    
    def dibujar_muchas_fuerzas(self):
        
        """
        Dibuja lo hecho por el método muchas_fuerzas()
        """
        
        fig, ax = plt.subplots(1,2, figsize = (15,7), dpi = 80)
        
        ax[0].set_title("%d Fuerzas" % len(self.vector_fuerzas), fontsize = 20)
        ax[0].set_ylabel("Posicion Y (m)")
        ax[0].set_xlabel("Posicion X (m)")
        
        data_aux = 0
        for i in range(len(self.vector_fuerzas)):
            ax[0].plot(self.posicion_x[data_aux : data_aux + self.vector_pasos[i]], 
                       self.posicion_y[data_aux : data_aux + self.vector_pasos[i]], 
                       label = "Fuerza %d" % (i+1), color = np.random.rand(4), linewidth = 2)
            ax[0].scatter(self.posicion_x[data_aux + self.vector_pasos[i]], 
                          self.posicion_y[data_aux + self.vector_pasos[i]], 30, color =  'y')
            data_aux += self.vector_pasos[i]
            
        ax[0].scatter(self.posicion_x[0], self.posicion_y[0], 40, label = "Pos inicial", c = 'b')
        ax[0].scatter(self.posicion_x[len(self.posicion_x)-1], 
                    self.posicion_y[len(self.posicion_y)-1], 40,  label = "Pos final", c = 'k')
        
        ax[0].legend(loc="best")
        
        
        ax[1].set_title("Movimiento continuo", fontsize = 20)
        ax[1].set_ylabel("Posicion Y (m)")
        ax[1].set_xlabel("Posicion X (m)")
        
        ax[1].plot(self.posicion_x, self.posicion_y, color = 'g', linewidth = 1.5)
            
        ax[1].scatter(self.posicion_x[0], self.posicion_y[0], 40, label = "Pos inicial", c = 'b')
        ax[1].scatter(self.posicion_x[len(self.posicion_x)-1], 
                    self.posicion_y[len(self.posicion_y)-1], 40,  label = "Pos final", c = 'k')
        
        ax[1].legend(loc="best")
```

```python
class Universo:
    
    def __init__(self, radio, N, particulas):
        self.radio = radio
        self.N = N
        self.particulas = particulas
        self.posiciones = []
        self.velocidades = []
        self.fuerzas = np.zeros([self.N,2]) 
        
    def fuerza_total(self):
        """
        Calcula la fuerza general que ejerce el sistema sobre cada partícula
        """
        
        for i in range(self.N):
            fuerza_aux_x = 0
            fuerza_aux_y = 0
            
            for j in range(self.N):
                fuerza_aux_x += self.particulas[i].fuerzaAplicada(self.particulas[j])[0]
                fuerza_aux_y += self.particulas[i].fuerzaAplicada(self.particulas[j])[1]
            
            self.fuerzas[i][0] = fuerza_aux_x
            self.fuerzas[i][1] = fuerza_aux_y
            

    
    def step(self, cambio_tiempo, subpasos):
        
        """
        Calcula el cambio de estado de todas las particulas en un lapso cambio_tiempo = dt
        """
        
        for i in range(self.N):
            self.particulas[i].mover(self.fuerzas[i], cambio_tiempo, subpasos) 
            
            
    def simular(self, tiempo, pasos, subpasos):
        """
        Simula los N cuerpos en un tiempo dado
        Usa el método step pasos veces
        """
        dt = tiempo/pasos
        for i in range(pasos):
            self.step(dt, subpasos) 
            self.fuerza_total()# Hecho el step de dt segundos se vuelven a reasignar las fuerzas
            
    def dibujar_universo(self):
        plt.figure(figsize = (10,10), dpi = 80)
        plt.xlim(-self.radio, self.radio)
        plt.ylim(-self.radio, self.radio)
        for i in range(0, int(self.N)):
            self.particulas[i].movimiento_puntual()
            plt.plot(self.particulas[i].posicion_x, self.particulas[i].posicion_y, c = np.random.rand(self.N), label = 
                     u"Partícula %d" % (i + 1))
            plt.scatter(self.particulas[i].scatter_x, self.particulas[i].scatter_y, 30,  
                        label = u"Movimiento puntual Partícula %d" % (i + 1), c = np.random.rand(self.N))
            
        plt.legend(loc = "best")
        
```
